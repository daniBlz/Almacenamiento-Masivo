```python
import requests
from lxml import html
import pandas as pd
```

```python
language = 'es'
pageid = 1224915242285
tipoPublicacion = 'Contratos Menores'
fechaPublicDesde = '01/01/2022'
pagename = 'PortalContratacion/Comunes/Presentacion/PCON_resultadoBuscadorAvanzado'
pags_total = 2
url_base = 'http://www.madrid.org/'
```

```python
titulo = ['Estado de la licitación','Resultado','Tipo Publicación','Objeto del contrato','Núm Expediente',
         'Referencia','Tipo de contrato','Procedimiento Adjudicación','Fecha del contrato',
         'Fecha de publicación','Duración del contrato','Número ofertas','Entidad Adjudicadora','NIF','nombre_ss','importe_sin_iva',
         'importe_con_iva']

estado_licitacion = []
resultado = []
tipo_publicacion = []
objeto_contrato = []
num_expediente = []
referencia = []
tipo_contrato = []
procedimiento_adjudicacion = []
fecha_contrato = []
fecha_publicacion = []
duracion_contrato = []
num_ofertas = []
entidad_adjudicadora = []
nif = []
nombre_ss=[]
importe_sin_iva = []
importe_con_iva = []

variables =[estado_licitacion,resultado,tipo_publicacion,objeto_contrato,num_expediente,referencia,tipo_contrato
,procedimiento_adjudicacion,fecha_contrato,fecha_publicacion,duracion_contrato,num_ofertas,entidad_adjudicadora,
           nif,nombre_ss,importe_sin_iva,importe_con_iva]
```

```python
url = 'http://www.madrid.org/cs/Satellite'

s = requests.Session()
```

```python
def info(url):
    rep2 = s.get(url = url,headers={'User-Agent': 'Mozilla/5.0'})
    html_tree2 = html.fromstring(rep2.text)
    info = html_tree2.xpath("//*[@id='cont_int_izdo']/div/ul/li/text()")
    info = list(filter(('\n').__ne__, info))
    info = list(filter((' ').__ne__, info))
    enlace = html_tree2.xpath('//*[@id="cont_int_izdo"]/div/ul/li/a[3]/text()')
    info.append(enlace[0])
    tabla = html_tree2.xpath('//*[@id="cont_int_izdo"]/div/ul/li[12]/table/tbody/tr/td/text()')
    for dato in tabla:
        info.append(dato)
    for i in range(0,17):
    
        variables[i].append(info[i])
        
    
```

```python
def iterar_href(enlaces):
    for enlace in enlaces:
        enlace = (url_base+enlace)
        info(enlace)
```

```python
def scrap(url):
    for i in range(1,pags_total+1):
        
        formdata = {'language':'es',
        'pageid':'1224915242285',
        'tipoPublicacion':'Contratos Menores',
        'fechaPublicDesde':'01/01/2022',
        'pagename':'PortalContratacion/Comunes/Presentacion/PCON_resultadoBuscadorAvanzado',
           'newPagina':i}
        req = s.post(url, data = formdata)
        html_tree = html.fromstring(req.text)
        enlaces = html_tree.xpath('//*[@id="fondoDoc"]//div[@class ="cajaBlanca"]//a/@href')
        iterar_href(enlaces)
    dicc = dict(zip(titulo,variables))
    df = pd.DataFrame.from_dict(dicc)
    return df
    
```

```python
df = scrap(url)
```
